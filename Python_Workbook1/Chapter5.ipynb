**CHALLENGE 1**

```python
total_sum = 0
count = 0
numbers = []

while True:
    user_input = input("Enter a number: ")

    if user_input.lower() == 'done':
        break

    try:
        num = float(user_input)
        total_sum += num
        count += 1
        numbers.append(num)
    except ValueError:
        print("Invalid input")
        continue

if count > 0:
    average = total_sum / count
else:
    average = 0

print(f"Total: {total_sum} Count:{count} Average: {average}")
```

**CHALLENGE 2**

```python
numbers = []

while True:
    user_input = input("Enter a number : ")

    if user_input.lower() == 'done':
        break

    try:
        num = float(user_input)
        numbers.append(num)
    except ValueError:
        print("Invalid input")
        continue

if len(numbers) == 0:
    print("No numbers were entered.")
else:
    maximum = max(numbers)
    minimum = min(numbers)

    print(f"Maximum: {maximum}")
    print(f"Minimum: {minimum}")
```
